### Question: Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.
##### Difficulty: Easy

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        hash_set = []

        for n in nums:
            if n not in hash_set:
                hash_set.append(n)

        if len(hash_set) != len(nums):
            return True
        else:
            return False
```

```python
solution = Solution()

nums = [1, 2, 3, 3]
result = solution.hasDuplicate(nums)
print(result)
```
