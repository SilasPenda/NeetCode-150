### Question: You are given an array of non-negative integers height which represent an elevation map. Each value height[i] represents the height of a bar, which has a width of 1.

### Return the maximum area of water that can be trapped between the bars.

##### Difficulty: Hard

```python
from typing import List

class Solution:
    def trap(self, height: List[int]) -> int:
        max_area = 0
        for i in range(len(height)):
            if len(height) < 2:
                return max_area
            if i == 0:
                left_max_h = 0
                right_max_h = max(height[i+1:])
            elif i == len(height) - 1:
                right_max_h = 0
                left_max_h = max(height[:i])
            else:
                left_max_h = max(height[:i])
                right_max_h = max(height[i+1:])

            area = min(left_max_h, right_max_h) - height[i]
            if area < 0:
                area = 0
                
            max_area += area

        return max_area
            
```

```python
solution = Solution()

height = [0,2,0,3,1,0,1,3,2,1]
result = solution.trap(height)
print(result)
```
