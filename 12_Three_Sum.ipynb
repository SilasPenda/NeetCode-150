### Question: Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] where nums[i] + nums[j] + nums[k] == 0, and the indices i, j and k are all distinct.

### The output should not contain any duplicate triplets. You may return the output and the triplets in any order.

##### Difficulty: Medium

```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        target = 0
        three_sum_list = []
        
        for i in range(len(nums)):    
            left = i + 1
            right = len(nums) - 1

            if i > 0 and nums[i] == nums[i - 1]:
                continue

            while left < right:
                total = nums[left] + nums[i] + nums[right]
 
                if total < target:
                    left += 1
                elif total > target:
                    right -= 1
                else:
                    three_sum_list.append([nums[left], nums[i], nums[right]])
                    left += 1

                    while nums[left] == nums[left - 1] and left < right:
                        left += 1

        return three_sum_list
```

```python
solution = Solution()

nums = [-1,0,1,2,-1,-4]
result = solution.threeSum(nums)
print(result)
```
