### Question: Design an algorithm to encode a list of strings to a string. The encoded string is then sent over the network and is decoded back to the original list of strings.

##### Difficulty: Medium

```python
from typing import List

class Solution:
    def encode(self, strs: List[str], delimiter: str=":;?") -> str:
        # Scenario 1: Empty list
        if len(strs) == 0:
            return "empty"

        # Scenario 2: List with strings (can be empty)
        encoded_strs = delimiter.join(strs)

        return encoded_strs

    def decode(self, s: str, delimiter: str=":;?") -> List[str]:
        if s == "empty":
            return []

        decoded_strs = s.split(delimiter)

        return decoded_strs
```

```python
solution = Solution()

inputs = ["Hello","World"]
result = solution.encode(inputs)
print(result)
```
