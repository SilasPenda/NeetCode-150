### Question: Given an integer array nums, return an array output where output[i] is the product of all the elements of nums except nums[i].

### Each product is guaranteed to fit in a 32-bit integer.

### Follow-up: Could you solve it in 0(n) time without using the division operation?

##### Difficulty: Medium

```python
from typing import List

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        # Scenarion 1: Empty list or single value list
        len_nums = len(nums)
        if len_nums < 1:
            return []

        # Scenarion 2: Multi-value list
        output_list = [1] * len_nums

        prefix = 1
        for i in range(len_nums):
            output_list[i] = prefix 
            prefix = output_list[i] * nums[i]

        suffix = 1
        for i in range(len_nums - 1, -1, -1):
            output_list[i] *= suffix
            suffix = suffix * nums[i]

        return output_list
```

```python
solution = Solution()

nums = [1,2,4,6]
result = solution.productExceptSelf(nums)
print(result)
```
