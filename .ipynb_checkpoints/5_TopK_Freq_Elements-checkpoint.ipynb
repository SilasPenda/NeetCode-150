### Quesiton: Given an integer array nums and an integer k, return the k most frequent elements within the array.

### The test cases are generated such that the answer is always unique.

### You may return the output in any order.

##### Difficulty: Medium

```python
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        set_nums = set(nums)

        nums_dict = {}

        for n in set_nums:
            num_list = [num for num in nums if num == n]

            nums_dict[n] = len(num_list)
            sorted_keys = sorted(nums_dict, key=nums_dict.get, reverse=True)

        return sorted_keys[:k]
```

```python
solution = Solution()

nums, k = [1,2,2,3,3,3], 2
result = solution.topKFrequent(nums, k)
print(result)
```
