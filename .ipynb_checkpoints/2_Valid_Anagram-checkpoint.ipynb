### Question: Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false. An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

##### Difficulty: Easy

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        s_count = {}
        t_count = {}

        for i in range(len(s)):
            s_count[s[i]] = s_count.get(s[i], 0) + 1
            t_count[t[i]] = t_count.get(t[i], 0) + 1

        for char in s_count:
            if s_count[char] != t_count.get(char, 0):
                return False

        return True
```

```python
solution = Solution()

s, t = "racecar", "carrace"
result = solution.isAnagram(s, t)
print(result)
```
